# fer

### <a href=https://towardsdatascience.com/the-ultimate-guide-to-emotion-recognition-from-facial-expressions-using-python-64e58d4324ff> The Code Reference <a/>

This code uses the FER library

- The Face Emotion Recognizer (generally knowns as the FER) is an open-source Python library built and maintained by Justin Shenk
- Used for sentiment analysis of images and videos.
- Built on a version that uses a convolution neural networks

```python
pip install FER
```

```python
from fer import FER
import matplotlib.pyplot as plt 
import cv2
```

```python
image = plt.imread("img/happy-1.jpg")
emo_detector = FER(mtcnn=True)


# Capture all the emotions on the image
captured_emotions = emo_detector.detect_emotions(image)


# Print all captured emotions with the image
print(captured_emotions)
plt.imshow(image)

# Use the top Emotion() function to call for the dominant emotion in the image
dominant_emotion, emotion_score = emo_detector.top_emotion(image)
print(dominant_emotion, emotion_score)
```

```python
#import the required modules
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace
```

```python
# read image
img = plt.imread('img/happy.png')
# img = img[800:2400, 900:2000, : :]
  
# call imshow() using plt object
plt.imshow(img)

# display that image
plt.show()

# storing the result
result = DeepFace.analyze(img, actions = ['emotion'], silent=True)
  
# print result
result[0]
```
